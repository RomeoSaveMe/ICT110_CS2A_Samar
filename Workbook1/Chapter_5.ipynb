```python
Enter a number: 4
Enter a number: 5
Enter a number: bad data
Invalid input
Enter a number: 7
Enter a number: done
16 3 5.333333333333333
```

```python
##answer
total = 0
count = 0

while True:
    num_input = input("Enter a number: ")

    if num_input == 'done':
        break

    try:
        num = float(num_input)
        total += num
        count += 1
    except ValueError:
        print("Invalid input")

if count == 0:
    print("No numbers entered.")
else:
    average = total / count
    print(total, count, average)
```

```python
Write another program that prompts for a list of numbers as above and at the end prints out both the maximum and minimum of the numbers instead of the average.
```

```python
numbers = []

while True:
    num_input = input("Enter a number: ")

    if num_input == 'done':
        break

    try:
        num = float(num_input)
        numbers.append(num)
    except ValueError:
        print("Invalid input")

if not numbers:
    print("No numbers entered.")
else:
    max_num = max(numbers)
    min_num = min(numbers)
    print("Maximum:", max_num)
    print("Minimum:", min_num)
```
