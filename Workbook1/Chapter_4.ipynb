```python
def com_pay(hours, rate):
    if hours <= 40:
        pay = hours * rate
    else:
        regular_hours = 40
        overtime_hours = hours - 40
        pay = (regular_hours * rate) + (1.5 * overtime_hours * rate)
    return pay

hours = float(input("Enter Hours: "))
rate = float(input("Enter Rate: "))

pay = com_pay(hours, rate)
print("Pay:", pay)
```

```python
Score           Grade
[ 97, 100]      1.0
[ 93,  97)        1.25
[ 89,  93)        1.5
[ 85,  89)        1.75
[ 81,  85)        2.0
[ 77,  81)        2.25
[ 74,  77)        2.5
[ 0,   74)        3.0

Run the program repeatedly to test the various different values for input.
```

```python
def computegrade(score):
    if score >= 97:
        return '1.0'
    elif score >= 93:
        return '1.25'
    elif score >= 89:
        return '1.5'
    elif score >= 85:
        return '1.75'
    elif score >= 81:
        return '2.0'
    elif score >= 77:
        return '2.25'
    elif score >= 74:
        return '2.5'
    elif score >= 0:
        return '3.0'
    else:
        return 'Score is out of range.'

score = float(input("Enter the score: "))

print("Grade:", computegrade(score))
```
